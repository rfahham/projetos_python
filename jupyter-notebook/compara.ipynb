```python
# Compara duas strings
# Neste caso tudo que vem depois do HTTP vai dar erro

string1 = "http://www.globo.com"

string2 = "https://www.globo.com"

conjunto = enumerate(zip(string1, string2))
# print(list(conjunto))

for indice, (a, b) in conjunto:
    if a != b:
        print(f"Item errado: {indice}")
```

```python
elocidade = float(input('Qual é a velocidade atual do carro? '))
if velocidade > 80:
	print('\033[31mMULTADO!\033[m Você excedeu o limite permitido que é de 80 Km/h' )
	multa = (velocidade - 80) * 7
	print('Você deve pagar uma multa de\033[31m R$ {:.2f}\033[m'.format(multa))
print('Tenha um bom dia, dirija com segurança!')

```

```python
import time
from itertools import cycle

cores = {
			'clear':'\033[m',
			'red':'\033[31m',
			'green':'\033[32m',
			'yellow':'\033[33m',
			'azulescuro':'\033[34m',	
			'roxo':'\033[35m',
			'azulclaro':'\033[36m',	
			'pretoebranco':'\033[7;30m',	
        }

lights = [('Green', 2),('Yellow', 0.5),('Red', 2)]

colors = cycle(lights)

while True:
    c,s = next(colors)
    print(c)
    time.sleep(s)
```
