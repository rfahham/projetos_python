```python
from bcb import sgs
import matplotlib.pyplot as plot

cdi = sgs.get({'CDI': 4390}, start='2024-01-01')

cdi.head()

cdi['CDI'] = cdi['CDI']/100

cdi['CDI_acum'] = (1 + cdi['CDI']).cumprod()

cdi['CDI_acum'].plot()
```
