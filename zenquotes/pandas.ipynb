```python
import requests
import pandas as pd

# URL da API
url = "https://zenquotes.io/api/quotes"

# Fazer a requisição para a API
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Carregar o conteúdo JSON
    quotes = response.json()
    
    # Criar um DataFrame a partir dos dados
    df = pd.DataFrame(quotes)
    
    # Exibir o DataFrame
    print(df)
else:
    print(f"Falha na requisição: {response.status_code}")
```

Listar somente a primeira coluna

```python
import requests
import pandas as pd

# URL da API
url = "https://zenquotes.io/api/quotes"

# Fazer a requisição para a API
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Carregar o conteúdo JSON
    quotes = response.json()
    
    # Criar um DataFrame a partir dos dados
    df = pd.DataFrame(quotes)
    
    # Listar apenas a primeira coluna (citações)
    primeira_coluna = df.iloc[:, 0]  # ou df['q'] para a coluna específica
    print(primeira_coluna.to_string(index=False))
else:
    print(f"Falha na requisição: {response.status_code}")
```

```python
import requests
import pandas as pd

# URL da API
url = "https://zenquotes.io/api/quotes"

# Fazer a requisição para a API
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Carregar o conteúdo JSON
    quotes = response.json()
    
    # Criar um DataFrame a partir dos dados
    df = pd.DataFrame(quotes)
    
    # Ajustar a configuração do pandas para exibir todas as linhas e sem compressão
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_colwidth', None)  # Permite colunas de largura ilimitada
    
    # Listar apenas a coluna de citações
    print(df['q'].to_string(index=False))
else:
    print(f"Falha na requisição: {response.status_code}")
```
